```python
#importing thr libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# loadind the data set
fifa = pd.read_csv("Downloads/players_20.csv")
```

```python
fifa.head() #getting the first 5 records of the data
```

```python
fifa.info() #checking the basic information of the data
```

```python
fifa.shape #checking the total number of rows and columns in the dataset
```

```python
for col in fifa: #printing all the names of the columns of the data
    print(col)
```

```python
fifa["nationality"].value_counts()[0:5] #finding the top 5 countries with the highest number of players 
```

```python
# creating a bar plot for top 5 counties with highest number of players
plt.figure(figsize=(8,5))
plt.bar(list(fifa["nationality"].value_counts()[0:5].keys()),list(fifa["nationality"].value_counts()[0:5]),color='purple')

for i,j in enumerate(list(fifa["nationality"].value_counts()[0:5])): #labels on bar
    plt.text(i,j+1, str(j),ha='center',va='bottom')

plt.title('Top 5 countries with most no. of players')    
plt.xlabel("Countries---->")  
plt.ylabel("No. of Players----->")
plt.show()
```

```python
player_salary = fifa[["short_name","wage_eur"]] #creating a new dataframe for players with their salary
```

```python
player_salary.head() #printing the first 5 names of the new data 
```

```python
player_salary = player_salary.sort_values(by= "wage_eur", ascending = False) #sorting the data of new df in decending order of the wages
```

```python
player_salary.head() # first 5 of the data
```

```python
# creating a bar plot of the data of top 5 highest paid players
plt.figure(figsize=(7,7))
plt.bar(list(player_salary["short_name"][0:5]), list(player_salary["wage_eur"][0:5]), color = "pink")

for i, j in enumerate(list(player_salary["wage_eur"][0:5])):
    plt.text(i, j+1, str(j), ha="center", va="bottom")

plt.title("Top 5 highest paid players")    
plt.xlabel("Players---->")
plt.ylabel("Salary----->")
plt.show()    
```
